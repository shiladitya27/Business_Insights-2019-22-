```python
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = data.dropna(subset=['October', 'November', 'June', 'July', 'August', 'September',
          'marketing', 'price', 'comp_marketing', 'macro', 'promo', 'comp_promo'])
```

```python
X = data[['October', 'November', 'June', 'July', 'August', 'September',
          'marketing', 'price', 'comp_marketing', 'macro', 'promo', 'comp_promo']]
y = data['sales']
```

```python
X = sm.add_constant(X)
```

```python
model = sm.OLS(y, X).fit()
```

```python
sm.OLS(y, X).fit().summary()
```

```python
predicted_sales = model.predict(X)
```

```python
# Calculate Durbin-Watson statistic
durbin_watson_stat = round(sm.stats.stattools.durbin_watson(model.resid),2)
print(f"Durbin-Watson Statistic: {durbin_watson_stat}")

# Calculate MAPE (Mean Absolute Percentage Error)
mape = np.mean((y - predicted_sales) / y) * 100
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")

# R-squared
r_squared = model.rsquared
print(f"R-squared: {r_squared:.2f}")
```

```python
import plotly.graph_objects as go
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=data.index,
    y=y,
    mode='lines+markers',
    name='Actual Sales',
    line=dict(color='blue'),
    marker=dict(symbol='circle')
))

fig.add_trace(go.Scatter(
    x=data.index,
    y=predicted_sales,
    mode='lines+markers',
    name='Predicted Sales',
    line=dict(color='red', dash='dash'),
    marker=dict(symbol='x')
))

fig.update_layout(
    title='Actual vs Predicted Sales',
    xaxis_title='Time',
    yaxis_title='Sales',
    legend=dict(x=0, y=1),
    grid=dict(
        rows=True,
        columns=True
    )
)

fig.show()
```

```python
residuals = y - predicted_sales
```

```python
fig = px.scatter(x=data.index, y=residuals,
                 title="Residuals vs Time",
                 labels={'x': 'Time', 'y': 'Residuals'})

fig.add_hline(y=0, line_dash="dot", line_color="red", line_width=1.2)
fig.show()
```

```python
fig = px.scatter(x=predicted_sales, y=residuals,
                 title="Residuals vs Predicted Values",
                 labels={'x': 'Predicted Values', 'y': 'Residuals'})

fig.add_hline(y=0, line_dash="dot", line_color="red", line_width=1.2)
fig.show()
```

```python
fig = px.histogram(residuals, nbins=30,
                   title="Histogram of Residuals",
                   labels={'value': 'Residuals', 'count': 'Frequency'})

fig.update_traces(marker_color="blue", marker_line_color="black",
                  marker_line_width=1)

fig.show()
```

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt

def qq_plot_residuals(residuals):
    """
    Creates a Q-Q plot for residuals to check for normality.

    Parameters:
    - residuals (pd.Series or np.array): The residuals from the model.
    """
    plt.figure(figsize=(10, 6))
    sm.qqplot(residuals, line='s', ax=plt.gca())
    plt.title("Q-Q Plot of Residuals")
    plt.grid(True)
    plt.show()

qq_plot_residuals(residuals)
```
